## Section 1 - Background on Quantum Computer Networks

Quantum computer networks are an emerging technology that leverages the principles of quantum mechanics to enable secure and efficient communication between quantum computers. These networks are composed of several key components in its architecture.

### Architecture

The architecture of a quantum computer network typically consists of quantum nodes, which are the quantum computers or quantum processors that perform computations and store quantum information. These nodes are connected by quantum channels, which are used to transmit quantum information between nodes. Quantum repeaters are used to extend the range of quantum communication by amplifying and correcting quantum signals. [1]

### Key Components

1. **Quantum Nodes**: These are the fundamental units of a quantum network, responsible for processing and storing quantum information.
2. **Quantum Channels**: These are the communication links that transmit quantum information between quantum nodes. They can be optical fibers or free-space links.
3. **Quantum Repeaters**: These devices are used to extend the communication range by overcoming the loss and decoherence of quantum signals over long distances.

### Significance

Quantum computer networks hold significant promise for the future of secure communication and distributed quantum computing. They enable the implementation of quantum key distribution (QKD) protocols, which provide theoretically unbreakable encryption. Additionally, quantum networks facilitate the sharing of quantum resources and the execution of distributed quantum algorithms, paving the way for advancements in quantum computing and communication technologies. [2]

### Coexistence with Classical Networks

It is important to note that quantum networks will not replace classical networks. Instead, they will coexist alongside classical networks, complementing them by providing enhanced security and new capabilities that classical networks cannot offer. Classical networks will continue to handle traditional data communication, while quantum networks will be used for specific applications that require the unique properties of quantum mechanics.

### Stages of Quantum Network Development

**Stage 1**: Any network is built on top of point-to-point connections. The initial stage of quantum network development will be defined by technology that enables a quantum analog of the Physical layer of the networking stack, where entanglement can be established between two separate quantum devices.

**Stage 2**: As there are limitations to scaling point-to-point connections, the next stage of quantum networking will be defined by technology that enables the analog of a Link layer. At this stage, a quantum device can support and manage connections with many sites, delivering entanglement to any two as required.

**Stage 3**: The final stage of development will be characterized by technology that enables a Network layer for reliable long-distance quantum communication through a complex network, which relies on resilient quantum hardware at the sites. [2]

## Section 2 - Overview of Practical Entanglement Distillation

Entanglement distillation is a crucial process in quantum communication that aims to enhance the quality of entangled quantum states. It involves taking several imperfect or noisy entangled states and transforming them into a smaller number of higher-fidelity entangled states. This process is essential for reliable quantum communication over long distances, where entanglement can degrade due to noise and other environmental factors.

### Concept of Entanglement Distillation

Entanglement distillation is based on the principles of quantum mechanics, specifically the phenomenon of entanglement, where two or more quantum particles become interconnected in such a way that the state of one particle directly influences the state of the other, regardless of the distance between them. However, in practical scenarios, entangled states are often subject to decoherence and noise, which reduce their fidelity and usefulness for quantum communication. [3]

### Role in Quantum Communication

In quantum communication, high-fidelity entangled states are essential for various protocols, including quantum key distribution (QKD), quantum teleportation, and distributed quantum computing. Entanglement distillation ensures that the entangled states used in these protocols are of sufficient quality to guarantee secure and efficient communication. By improving the fidelity of entangled states, entanglement distillation enhances the overall performance and reliability of quantum networks.

### Practical Techniques for Enhancing Entanglement Fidelity

Several practical techniques have been developed to perform entanglement distillation, including:

1. **Entanglement Purification Protocols**: These protocols, such as the DEJMPS protocol, use local operations and classical communication (LOCC) to iteratively improve the fidelity of entangled states. They typically involve multiple rounds of operations and measurements to filter out lower-fidelity states.

2. **Error Correction Codes**: Quantum error correction codes can be used to protect entangled states from noise and decoherence. By encoding the entangled states into larger quantum systems, errors can be detected and corrected, thereby improving the overall fidelity.

3. **Entanglement Swapping**: This technique involves creating entanglement between distant particles by performing joint measurements on intermediate particles that are entangled with the distant ones. Entanglement swapping can be combined with distillation protocols to extend the range of high-fidelity entanglement.


## Section 3 - Quantum vs. Classical Communication Networks

Quantum communication networks and classical communication networks differ fundamentally in their underlying principles and capabilities. Some critical differences include:

### Entanglement

- **Quantum Networks**: Utilize entanglement, a unique quantum phenomenon where particles become interconnected, and the state of one particle instantly influences the state of another, regardless of distance. This property is crucial for quantum communication protocols.
- **Classical Networks**: Do not have an equivalent to entanglement. Information is transmitted through classical signals, which are subject to the speed of light and can be intercepted or tampered with.

### Superposition

- **Quantum Networks**: Exploit superposition, where quantum bits (qubits) can exist in multiple states simultaneously. This allows for more complex and efficient information processing and transmission.
- **Classical Networks**: Use classical bits that are either in a state of 0 or 1. This binary nature limits the amount of information that can be processed and transmitted at any given time.

### Quantum Key Distribution (QKD)

- **Quantum Networks**: Enable Quantum Key Distribution, a secure communication method that uses quantum mechanics to generate and distribute cryptographic keys. QKD provides theoretically unbreakable encryption, as any attempt to eavesdrop on the key exchange will disturb the quantum states and be detected.
- **Classical Networks**: Rely on classical cryptographic methods, which are based on mathematical algorithms. These methods can be vulnerable to advances in computing power and cryptanalysis techniques.

### Error Correction

- **Quantum Networks**: Require quantum error correction to protect against decoherence and other quantum noise. Quantum error correction codes are more complex than classical ones, as they must account for the unique properties of quantum information.
- **Classical Networks**: Use classical error correction techniques to detect and correct errors in data transmission. These techniques are well-established and effective for classical information.

In conclusion, quantum communication networks offer enhanced security and new capabilities through the principles of entanglement and superposition. They enable secure key distribution and require advanced error correction methods. In contrast, classical communication networks rely on traditional cryptographic methods and binary data transmission, which are more susceptible to interception and computational advances.

## Section 4 - Methodology for Formulating an Optimization Model for Entanglement Distillation
For this lab, the proposed methodology works over the concept of Local Operations and Classical Communication (LOCC) and the developments of previously proposed distillation protocols such as BBPSSW and DEJMPS. First, it is important to identify the state fidelity $ F $ between the distilled state $ \rho_{\text{out}} $ and the maximally entangled Bell state  $ \lvert \Phi^+ \rangle $ to quantify the performance of a distillation protocol, where:

$$
F(\rho_{\text{out}}, \Phi^+) \equiv \langle \Phi^+ \lvert \rho_{\text{out}} \lvert \Phi^+ \rangle.
$$

Then, to start the protocol we begin with the creation of four identical isotropic states which are a parametrized mixed state consisting of $ \lvert \Phi^+ \rangle $ and a white noise mixed state $ I/4 $. The full isotropic state is defined as:

$$
\rho_{\text{iso}}(p) = p \lvert \Phi^+ \rangle \langle \Phi^+ \rvert + (1 - p) \frac{I}{4}, \quad p \in [0, 1]
$$

The isotropic states represent the initial state that will be distributed from the source to both Alice $ (A) $ and Bob $ (B) $. Each will hold four copies of entangled qubit pairs $ \rho A_{\text{0}} B_{\text{0}} $, $ \rho A_{\text{1}} B_{\text{1}} $, $ \rho A_{\text{2}} B_{\text{2}} $ and $ \rho A_{\text{3}} B_{\text{3}} $ with Alice holding qubits $ (A_{\text{0}}, A_{\text{1}}, A_{\text{2}}, A_{\text{3}}) $ in one location and Bob holding $ (B_{\text{0}}, B_{\text{1}}, B_{\text{2}}, B_{\text{3}}) $ in a separate location. At this point, if the measurement results of all qubit pairs are the same, meaning that $ m_{A_{\text{1}}} m_{B_{\text{1}}}, m_{A_{\text{2}}}m_{B_{\text{2}}}, m_{A_{\text{3}}}m_{B_{\text{3}}} $ is either $ 00 $ or $ 11 $ we can understand that the distillation is successful and qubits $ A_{\text{0}} $ and $ B_{\text{0}} $ will store the purified entanglement state $ \rho'_{AB} $. If this condition is not met, then the pair is discarded and the protocol is restarted.

Now, for the optimization of the distillation process we define a loss function in the form of:

$$
L = \sum_{m_{A_j} m_{B_j} \in \{00, 11\}} \left( 1 - \text{Tr}(\rho_{\text{target}} \rho'_{AB}) \right),
$$

where $ \text{Tr}(\rho_{\text{target}} \rho'_{AB}) $ is the state overlap between $ \rho'_{AB} $ and the target state:

$$
\rho_{\text{target}} = \lvert \Phi^+ \rangle \langle \Phi^+ \rvert
$$

Then, we use classical optimization methods to vary the QNN circuit parameters to minimize the loss function and maximize fidelity.

## Section 5 - Methodology Implementation Using Qiskit and Scipy
We start with the required imports for the project:

Then, we need to define the source for the qubits that will be distributed to Alice and Bob. For this we start by creating a function to prepare the bell state $ \lvert \Phi^+ \rangle $:

```python
# Function to create a Bell state between two qubits
def create_bell_state_cont(num_qubits):
    if num_qubits % 2 != 0:
        raise ValueError("Number of qubits must be even to create Bell states on each pair.")
    
    qc = QuantumCircuit()
    alice = QuantumRegister(num_qubits // 2, 'Alice')
    bob = QuantumRegister(num_qubits // 2, 'Bob')
    qc.add_register(alice)
    qc.add_register(bob)
    for i in range(0, num_qubits, 2):
        qc.h(i)  # Apply Hadamard gate to qubit i
        qc.cx(i, i+1)  # Apply CNOT gate with control=i and target=i+1
    return qc
```

With this implementation, we can continue and define a function that will add the white noise mixed state and the initial fidelity value to create an initial state based on the number of desired entangled pairs:

```python
# Function to create initial state for a number of qubit pairs and a given fidelity
def create_initial_state(num_pairs, prob):
    # Create a Bell state circuit with 2 qubits
    qc = create_bell_state_cont(2)
    
    # Define the dimension of the density matrix
    dim = 2 ** (2)

    # Convert the Bell state circuit to a density matrix
    bell_density_matrix = DensityMatrix(qc)
    
    # Create isotropic state with specified probability
    iso_state = prob * bell_density_matrix + (1 - prob) * np.eye(dim) / dim

    # Initialize the final matrix with the density matrix of the Bell state
    final_matrix = iso_state
    
    # Perform the Kronecker product of the density matrix with itself num_pairs-1 times
    for _ in range(num_pairs - 1):
        final_matrix = np.kron(final_matrix, iso_state)
    
    if num_pairs == 1:
        return final_matrix.data
    else:
        return final_matrix
```

For this experiment, we use the following values to create the initial state density matrix:

```python
prob = 0.7
num_pairs = 4
initial_state_matrix = create_initial_state(num_pairs, prob)
```

To continue, we need to define the QNN circuit. For this specific application, we need to define a parameter vector that will permit the assignment of specific values to the circuit during each iteration of the optimization process. We define the QNN circuit as follows:

```python
def nlocal_circuit(num_pairs, angles=None):

    num_layers = 1
    qc = QuantumCircuit()
    
    # Create registers for Alice and Bob
    for i in range(num_pairs):
        alice = QuantumRegister(1, f'A{i}')
        bob = QuantumRegister(1, f'B{i}')
        qc.add_register(alice)
        qc.add_register(bob)

    # Create NLocal Parametrized Circuit
    theta = Parameter("Î¸")
    ansatz_A = NLocal(
        num_qubits=num_pairs,
        rotation_blocks=[RXGate(theta), CRZGate(theta)],
        entanglement_blocks=CCXGate(),
        entanglement = [[0, 1, 2], [0, 2, 3], [3, 1, 0]],
        reps=num_layers,
        insert_barriers=True,
    )

    # Rename Bob circuit parameters to avoid conflicts
    gammas = ParameterVector("Bob", len(ansatz_A.parameters))
    ansatz_B = ansatz_A.assign_parameters(dict(zip(ansatz_A.parameters, gammas)))

    # Add QAOA Circuit to Alice Qubits
    qc.append(ansatz_A, [0,2,4,6])
    qc.barrier()

    # Add QAOA Circuit to BOB Qubits
    qc.append(ansatz_B, [1,3,5,7])

    return qc

# Initialize parameters array to random values
initial_angles = np.random.uniform(0, np.pi, 24)  # Adjusted for three parameters per qubit for two layers

test_qc = nlocal_circuit(4, initial_angles)
test_qc.draw('mpl')
```

Now, we can define the loss function callback following the details in Section 4. It is important to remark that we are using Qiskit density matrices in this implementation which is why we note the use of the **evolve()** method for applying the QNN circuit to the initial state. Also, the **measure()** method and the **partial_trace()** method are used to determine if the distillation process was successful to continue the protocol:

```python
from scipy.optimize import minimize

iterations = 0
loss_values = []

def loss_function(params, num_measurements=1000, desired_successful_cases=8):
    global iterations
    successful_cases = ["000000", "000011", "001100", "110000", "001111", "111100", "110011", "111111"]
    counts = 0
    loss = 0
    target_state = DensityMatrix(create_bell_state_cont(2))
    
    # Assign the parameters to the quantum circuit
    test_qc = qaoa_circuit(4, params)
    test_qc = test_qc.assign_parameters(params)
    
    # Initialize the state
    init_matrix = DensityMatrix(initial_state_matrix)
    init_matrix = init_matrix.evolve(test_qc)  # Apply the QNN circuit to the initial state

    init_matrix.seed(0)
    for _ in range(num_measurements):
        outcome, ret = init_matrix.measure([2, 3, 4, 5, 6, 7])
        if outcome in successful_cases:
            rho_ab = partial_trace(ret, [2, 3, 4, 5, 6, 7]).data
            loss += 1 - np.real(DensityMatrix(np.matmul(target_state.data, rho_ab)).trace())
            counts += 1
            if counts >= desired_successful_cases:
                break
    
    iterations += 1
    loss_values.append(loss)
    if iterations % 10 == 0:
        print("Iteration: ", iterations, " Loss: ", loss)
    return loss
```

With the loss function callback ready, we can use classical optimization solvers to minimize the loss function and obtain the results:

```python
# Using scipy optimizer
initial_angles = np.random.uniform(0, np.pi, 8)
result = minimize(loss_function, initial_angles, method='COBYLA', tol=1e-2, options={'maxiter': 2000, 'disp': False})
print("Optimized parameters:", result.x)
print("Minimum loss:", result.fun)
```
